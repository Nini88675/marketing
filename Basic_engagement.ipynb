```python
import pandas as pd

# Lire ton fichier CSV
df = pd.read_excel("videos_details.xlsx")

# V√©rifier que les colonnes existent
cols_needed = ['video_id', 'title', 'viewCount', 'likeCount', 'commentCount']
for c in cols_needed:
    if c not in df.columns:
        raise ValueError(f"Colonne manquante : {c}")

# Convertir en num√©rique au cas o√π certaines valeurs seraient lues comme texte
df['viewCount'] = pd.to_numeric(df['viewCount'], errors='coerce')
df['likeCount'] = pd.to_numeric(df['likeCount'], errors='coerce')
df['commentCount'] = pd.to_numeric(df['commentCount'], errors='coerce')
```

```python

# üìà Calcul du taux d'engagement (%)
df['engagement_rate_percent'] = (df['likeCount'] + df['commentCount']) / df['viewCount'] * 100

# üßæ Calcul des KPI globaux
kpi_data = {
    "KPI": [
        "Total Vues",
        "Total Likes",
        "Total Commentaires",
        "Engagement moyen (%)",
        "Vid√©o la plus vue (ID)",
        "Vid√©o avec le plus de likes (ID)",
        "Vid√©o avec le meilleur engagement (%) (ID)"
    ],
    "Valeur": [
        df['viewCount'].sum(),
        df['likeCount'].sum(),
        df['commentCount'].sum(),
        round(df['engagement_rate_percent'].mean(), 2),
        df.loc[df['viewCount'].idxmax(), 'video_id'],
        df.loc[df['likeCount'].idxmax(), 'video_id'],
        df.loc[df['engagement_rate_percent'].idxmax(), 'video_id']
    ]
}

# üìä Cr√©ation du DataFrame KPI
kpi_df = pd.DataFrame(kpi_data)
```
